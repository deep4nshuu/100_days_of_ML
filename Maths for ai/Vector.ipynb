```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
# plot with x & y limits with color
plt.quiver(0,0,4,6, scale_units='xy', angles='xy', scale=1, color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,6, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0,0,-3,-8, scale_units='xy', angles='xy', scale=1, color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

- Addition of two vectors

```python
vector_1 = np.asarray([0,0,4,2])
vector_2 = np.asarray([0,0,-3,2])

sum = vector_1 + vector_2
print(sum)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='y')
plt.quiver(0,0,3,-2, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,1,4, scale_units='xy', angles='xy', scale=1, color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

- Substraction of two vectors

```python
vector_1 = np.asarray([0,0,4,2])
vector_2 = np.asarray([0,0,-3,2])

sum = vector_1 - vector_2
print(sum)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='y')
plt.quiver(0,0,3,-2, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,7,0, scale_units='xy', angles='xy', scale=1, color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

- Multiplication by a scalar

```python
vector_1 = np.asarray([0,0,4,2])
vector_2 = -2 * vector_1
print(vector_2)
```

```python
# plot with x & y limits with color
plt.quiver(0,0,4,2, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0,0,-8,-4, scale_units='xy', angles='xy', scale=1, color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

- Dot product of two vectors

```python
a = np.array([3,5,6])
b = np.array([4,7,9])

a_dot_b = np.dot(a,b)
print(a_dot_b)
```

- Cross product of two vectors

```python
a = np.array([3,5,6])
b = np.array([4,7,9])

a_cross_b = np.cross(a,b)
print(a_cross_b)
```

- Projection of a vector on another vector

```python
a = np.array([3,5,6])
v = np.array([4,7,9])
# magnitude of v
mag_of_v = np.sqrt(np.sum(v**2))

proj_of_a_on_v = (np.dot(a,v)/mag_of_v**2)*v

print(proj_of_a_on_v)
```

- Element wise multiplication

```python
mat1 = np.random.randint(10,20, (3,3))
mat2 = np.random.randint(10,30, (3,3))

print(mat1)
print(mat2)
```

- When doing element wise multiply, shape should be same for both matrix
- But when doing multiplicatn of two matrix, then col of ist and row of 2nd should be same

```python
prod = np.multiply(mat1, mat2)
print(prod)
```
