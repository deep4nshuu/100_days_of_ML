# Classificatn Matrix
They are used to find the best or accurate model and help in comparing models 

> Various matices are:
- Accuracy
- Confusion matrix
- Precision
- Recall 
- F1 score

>a. Accuracy
    
        acc = No of correct pred / Total pred

- How much accuracy is good?

It depends on the problem AND the data we are using

- Problem with accuracy?

It only tell about mistake but doesn't tell about nature of mistake


> Confusion Matrix

<img src='imgA.png'></img>

It tells us about the nature of mistakes or what amt and what type of mistake been made

- Terminology(ActualPredtn): 11-TP(TruePositive), 10-FN(FalseNeg), 01-FP(FalsePos), 00-TN(TrueNeg)

- We can calculate Accuracy from Confusion Matrix but vice-versa not true

- Accuracy = (TP + TN) / (TP+TN+FP+FN)

- <p style='color:red'>Type1 error : Means FP(FalsePos) value</p>
- <p style='color:green'>Type2 error : Means FN(FalseNeg) value</p>

<img src='imgB.png'></img>

- When accuracy is misleading?

When Data is Imbalanced or ratio of true vs false is nearly 0

> Precision

It is the ratio of True Positive to the total Predicted Positive

<img src='imgC.png'></img>

Precision = TP / (TP + FP)

- If Type 1 error is high -> then select High Precision model

> Recall

What proportion of actual Positive is correctly classified

<img src='imgD.png'></img>

- If Type 2 error is dangerous -> then select High Recall model

> F1 score

Used when not able to differnetiate btw type 1 is more dangerous or type 2 is more

It is a combination of Precision and Recall

F1 score = 2 * P * R / (P + R)
