```python
import time

API_KEY = "c8ea7f1e2e3000adcb00db51a7b11fd9"
BASE_URL = "https://api.themoviedb.org/3/movie/top_rated"

session = requests.Session()  # 🔥 reuse connections
all_movies = []

for i in range(1, 500):
    try:
        response = session.get(
            BASE_URL,
            params={
                "api_key": API_KEY,
                "language": "en-US",
                "page": i
            },
            timeout=15
        )

        data = response.json()

        if "results" not in data:
            print(f"Stopped at page {i}: {data}")
            break

        temp_df = pd.DataFrame(data["results"])[
            ["id", "title","release_date", "overview", "popularity", "vote_average", "vote_count"]
        ]

        all_movies.append(temp_df)

        time.sleep(0.5)  # ⬅️ slower but STABLE

    except requests.exceptions.ConnectionError as e:
        print(f"Connection error at page {i}. Waiting & retrying...")
        time.sleep(5)
        continue

df = pd.concat(all_movies, ignore_index=True)
df
```

```python
df.to_csv('movies.csv')
```
