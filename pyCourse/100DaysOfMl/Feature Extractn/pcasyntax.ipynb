```python
from sklearn.metrics import accuracy_score
accuracy_score(y_ts, y_pred)
```

# **Now using PCA for this dataset** 

Step 1 : Mean Centric our data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_tr = scaler.fit_transform(x_tr)
x_ts = scaler.transform(x_ts)
```

**Applying PCA**

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=100)
```

```python
x_tr_trf = pca.fit_transform(x_tr)
x_ts_trf = pca.transform(x_ts)
```

```python
x_tr_trf.shape
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(x_tr_trf, y_tr)
```

```python
y_pred1 = knn.predict(x_ts_trf)
```

```python
accuracy_score(y_ts, y_pred1)
```

```python
for i in range(1,785):
    pca = PCA(n_components=i)
    x_tr_trf = pca.fit_transform(x_tr)
    x_ts_trf = pca.transform(x_ts)

    knn = KNeighborsClassifier()

    knn.fit(x_tr_trf, y_tr)

    y_pred1 = knn.predict(x_ts_trf)

    print(accuracy_score(y_ts, y_pred1))
```
