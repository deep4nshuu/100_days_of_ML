```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv(r'C:\Users\HP\OneDrive\Desktop\python\pyCourse\csvFiles\titanic\train.csv', usecols=['Age', 'Pclass', 'SibSp', 'Parch', 'Survived'])
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
x = df.iloc[:,1:5]
y = df['Survived']
```

```python
np.mean(cross_val_score(LogisticRegression(),x,y, scoring='accuracy', cv=20))
```

# Feature Construction

```python
x['Family_Size'] = x['SibSp']+ x['Parch'] + 1
```

```python
x.head()
```

```python
def myFn(num):
    if num == 0:
        # alone
        return 0
    elif num == 1:
        # small family
        return 1
    else :
        # large family
        return 2
```

```python
myFn(4)
```

Applying fn on dataset

```python
x['Family_Type'] = x['Family_Size'].apply(myFn)
```

```python
x.drop(columns=['SibSp','Parch','Family_Size'], inplace=True)
```

```python
x.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),x,y, scoring='accuracy', cv=20))
```
