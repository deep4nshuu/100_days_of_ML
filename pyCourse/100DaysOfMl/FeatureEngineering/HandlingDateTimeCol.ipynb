- when we import dataframe, it's date & time values are stored as objects(strings)
- So we need to change it into date & time

```python
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

Extracting year, month & month_name, date

```python
date['year'] = date['date'].dt.year
date['month'] = date['date'].dt.month
date['month_name'] = date['date'].dt.month_name()
date['day'] = date['date'].dt.day
date.head()
```

```python
# days of week
date['dow'] = date['date'].dt.day_of_week
# day of week name
date['dow_name'] = date['date'].dt.day_name()
# is weekend
date['date_is_weekend'] = np.where(date['dow_name'].isin(['Sunday','Monday']),1,0)
# week of year
date['date_week'] = date['date'].dt.weekday
# semester of year
date['semester'] = np.where(date['date'].isin([1,2]),1,2)
date.drop(columns=['product_id','city_id','orders']).head()
```

```python
import datetime
today = datetime.datetime.today()
today
```

```python
(today - date['date']).dt.days
```

```python
# month passed 
np.round((today - date['date']) / np.timedelta64(1,'W'),0)
```
