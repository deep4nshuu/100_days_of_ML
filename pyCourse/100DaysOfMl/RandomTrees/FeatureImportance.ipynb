# Calculating feature importance using DT

```python
from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier
```

```python
x,y = make_classification(n_samples=5, n_classes=2, n_features=2, n_informative=2, n_redundant=0, random_state=0)
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(x,y)
```

```python
from sklearn.tree import plot_tree
plot_tree(clf)
```

```python
clf.feature_importances_
```

# Calcuating using RF
It finds means of feature_importance calculated by individual DT's

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=2)
```

```python
rf.fit(x,y)
```

```python
rf.feature_importances_
```

```python
print(rf.estimators_[0].feature_importances_)
print(rf.estimators_[1].feature_importances_)

print((1 + 0.375)/2)
```
