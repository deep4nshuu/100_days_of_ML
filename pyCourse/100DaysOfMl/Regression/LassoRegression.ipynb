<h5>Sparsity</h5>

- when lambda inc and val of coef become zero, that condn is called sparisty
- So when val of lambda inc in Ridge -> coef moevs towards 0 -> but never become zero
- But in Lass, first some coef become 0 -> then gradually inc lambda -> all coef become 0

> Why it happen in Lasso?
<img src='img19.png'></img>
- When lambda inc but smaller than YX val -> then val of m is +ve but smaller than prev -> so m dec
- When lambda inc, equal to YX val -> then val of m become 0 -> but formula for m=0 cndn is used -> similar to SLR
- When lambda is unc but more than YX val -> m become -ve -> so formula for cndn(m<0) is used -> where m is added in YX which make it greater than prev vals -> which means coef should be -ve -> Our goal is to reduce m instead of it inc when m become negative
- So due to this -> As lambda inc -> val of coef become zero but not go beyond 0 to reduce confusion

> But why it does not happen in Ridge
- Beacause lambda is present in den -> so to make m 0 -> we need to make num 0 -> so there is no control of lambda on num to make it 0

# ElasticNet Regression
- It is the middle of Ridge and Lasso used for reduce overfitting when we don't know our cols are important or not
- Formula : L = sigma(Yi^2 - Yi'^2) + a|W^2| + b||W||
- It includes 2 penality terms for loss fn

> Used when Input cols have multi-colinarity means one depend on another
- In scikit when we apply EN, it provide 2 hyper params: lambda(a+b) and l1_ratio(a/(a+b))
- Def vals of HP are: lmbda=1 & l1_ratio=0.5

> In scikit, we can apply EN in two ways whenere one is by using EN directly or using SGDRegreesor where selecting penality as EN

```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
X,y = load_diabetes(return_X_y=True)
```

```python
x_tr, x_ts, y_tr, y_ts = train_test_split(X,y, test_size=0.2, random_state=2)
```

```python
# Linear Regression
lr = LinearRegression()
lr.fit(x_tr, y_tr)
y_pred = lr.predict(x_ts)
r2_score(y_ts, y_pred)
```

```python
# Ridge
rr = Ridge(alpha=0.1)
rr.fit(x_tr, y_tr)
y_pred = rr.predict(x_ts)
r2_score(y_ts, y_pred)
```

```python
# Lasso
lr = Lasso(alpha=0.01)
lr.fit(x_tr, y_tr)
y_pred = lr.predict(x_ts)
r2_score(y_ts, y_pred)
```

```python
# ElasticNet
en = ElasticNet(alpha=0.005, l1_ratio=0.9)
en.fit(x_tr, y_tr)
y_pred = en.predict(x_ts)
r2_score(y_ts, y_pred)
```
