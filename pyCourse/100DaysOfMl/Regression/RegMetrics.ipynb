> R2 score

<img src='img7.png'></img>
- it finds out how much LR line is better than mean line
- if 0, means SSr/SSm is 1 means LR line and mean line both are nearly same
- if 1, means SSr/SSm is 0 which means error fn is 0 or model is perfect
- Ex : R2 score is 0.80 of cgpa|lpa data which means cgpa explains 80% of variance in lpa

> Adjusted R2 score

<img src='img8.png'></img>
- Sometimes by adding features r2 score inc but sometimes features may be irrelevant and r2 score still inc which questions model accuracy
- So in this, if we add irrelevant cols, adj r2 score decreses
- Useful for Multiple LR or large dataset

# How to apply/find?

- import metrics from slearn.metrics
- now call them and pass y_test and y_pred
- mean_abs_error(y_test, y_pred)
- mean_sq_error(y_test, y_pred)
- for rmse : np.sqrt(mean_abs_error(y_test, y_pred))
- r2_score(y_test, y_pred)
