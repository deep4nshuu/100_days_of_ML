# Softmax Regression

A form of Logistic Regression which deals with more than two class data or Multinomial Logistic Reg

- LR is a specializatn of Softmax with 2 classes

> Softmax Fn

It calculate the probability of individual classes
<img src='img30.png'></img>

> Training Intuition
1. First transform third class(OHE)
2. Form 3 dataset with individual(OHE transformed) class
3. Treat them as 3 new model and find their coef which will be 3 for each
4. Then finding z val and then their prob

Conclusion -
- You are training Logistic Regression models equal to classes numbers and finding equal no of coef for each model
- Then finding z val for each model
- And then convert that z val into Probability using softmax fn and model with max prob, assign by pt

> Problem with this intuition
- Due to finding coef individually for each model, it makes process very slow
- And this intuition is not used by scikit learn
- SOLN : Scikit finds all of 9 coef with single model

> Implementatn

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
enc = LabelEncoder()
df['species'] = enc.fit_transform(df['species'])
```

```python
df.head()
```

```python
df = df[['sepal_length','petal_length','species']]
```

```python
df.head()
```

```python
x = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
x_tr, x_ts, y_tr, y_ts = train_test_split(x,y, test_size=0.2, random_state=2)
```

```python
clf = LogisticRegression(multi_class='multinomial')
```

```python
clf.fit(x_tr,y_tr)
```

```python
y_pred = clf.predict(x_ts)
```

```python
print(accuracy_score(y_ts,y_pred))
```

```python
pd.DataFrame(confusion_matrix(y_ts, y_pred))
```

```python
# predictn
query = np.array([3,4,2,7])
clf.predict_proba(query)
```

```python
clf.predict(query)
```

```python
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(x.values, y.values, clf, legend=2)

# Adding axes annotatn
plt.xlabel('sepal_length [cm]')
plt.ylabel('petal length [cm]')
plt.title('Softmax on Iris')

plt.show()
```
