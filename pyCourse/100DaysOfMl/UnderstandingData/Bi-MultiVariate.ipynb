<h3 style='color: brown'>2. Numerical-Categorical </h3>

<h4 style='color: pink'>a. Barplot </h4>

```python
sns.barplot(x=titanic['Pclass'], y=titanic['Age'], hue=titanic['Sex'])
```

<h4 style='color: pink'>b. Boxplot </h4>

```python
sns.boxplot(x=titanic['Sex'], y=titanic['Age'], hue=titanic['Survived'])
```

<h4 style='color: pink'>c. Distplot</h4>

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'], hist=False, color='Red')
sns.distplot(titanic[titanic['Survived']==1]['Age'], hist=False, color='Green')
```

<h3 style='color: brown'>3. Categorical-Categorical</h3>


<h4 style='color: pink'>a. Heatmap</h4>

```python
pd.crosstab(titanic['Pclass'], titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

<h4 style='color: pink'>b. ClusterMap</h4>

- Helps in finding hierarchial relatnships

```python
sns.clustermap(pd.crosstab(titanic['Parch'], titanic['Survived']))
```
