```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

<h2 style='color: brown'>1. Categorical data</h2>

- In categorical data, we can extract only frequency distributn, plot it
- Can plot -> countplot, piechart

<h3 style='color: pink'>a. Countplot</h3>

```python
sns.set()
```

```python
# sns.countplot(x='Survived', data=df)
df['Survived'].value_counts().plot(kind='bar')
```

```python
sns.countplot(x='Pclass', data=df)
```

<h3 style='color: pink'>b. piechart</h3>

```python
df['Pclass'].value_counts().plot(kind='pie', autopct='%.2f')
```

<h2 style='color: brown'>2. Numerical Data</h2>

- Plot distplot(for finding distributn), histogram(for visualizing distributn),etc
- Can plot -> histogram, distplot, boxplot

<h3 style='color: pink'>a. Histogram</h3>

- Helps in understanding distributn of data in particular col

```python
plt.hist(df['Age'], bins=10)
```

<h3 style='color: pink'>b. Distplot</h3>


- It is PDF(Probability Distr FN) helps in finding the probabiltiy of a value whereas histogram give only value range

```python
sns.distplot(df['Age'])
```

<h3 style='color: pink'>c. Boxplot</h3>

- It helps in identifying outliers in data  and noisy data

```python
sns.boxplot(x='Age', data=df)
```

<h3 style='color: pink'>d. Skewness</h3>

- Helps in finding skewness of data means it tells how much data is symmetrical on both sides
- If + -> means positively skewed toward right side
- If 0 -> means perfectly symmetrical
- If - -> means negatively skewed towards left

```python
df['Age'].skew()
```
