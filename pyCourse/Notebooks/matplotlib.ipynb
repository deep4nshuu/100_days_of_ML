# MATPLOTLIB

IT IS USEFUL FOR MAkING PLOTS AND GRAPHS

```python
# importing matplotlib lib
import matplotlib.pyplot as plt

# importing numpy to get data for our plots
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

# Plottting the data

.plot() fn   --> it will make empty plot

.show()   --> it will show the plot

```python
# sine wave
plt.plot(x,y)
plt.show()
```

```python
# cosine wave
plt.plot(x,z)
plt.show()
```

Adding title, x-axis & y-axis labels

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

# Constructing a parabola

```python
x = np.linspace(-10,10,20)
y = x**2   # x^2
plt.plot(x,y)
plt.xlabel('x values')
plt.ylabel('x^2/y values')
plt.title('Parabola')
plt.show()
```

```python
# Plotting line with diff symbol and color
plt.plot(x,y, 'r+')
plt.show()
```

Plotting Graph with multiple lines

```python
x = np.linspace(-5,5,50)
plt.plot(x, np.sin(x), 'r--')
plt.plot(x, np.cos(x), 'g-')
plt.show()
```

Plotting Bar Graph

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])   # ([origin,ht,wdt])
languages = ['English','Hindi', 'Spannish','Punjabi', 'Tamil']
people = [30,100,20,50,40]
ax.bar(languages, people)
plt.xlabel('LANGUAGE')
plt.ylabel('PEOPLE')
plt.show()
```

# A 3d bar graph

```python
plt.style.use("_mpl-gallery")
# Make data
x = [1,1,2,2]
y = [1,2,1,2]
z = [0,0,0,0]
dx = np.ones_like(x)*0.5
dy = np.ones_like(y)*0.5
dz = [2,3,1,4]

# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.bar3d(x,y,z,dx,dy,dz)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

# Pie Chart

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])   # ([origin,ht,wdt])
languages = ['English','Hindi', 'Spannish','Punjabi', 'Tamil']
people = [30,100,20,50,40]
ax.pie(people, labels=languages, autopct='%1.1f%%')   # autopct tell how much floating decimal to be taken
plt.show()
```

# Scatter plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])  # these dim are for 2d plot
ax.scatter(x,y, color='g')
ax.scatter(x,z, color='b')
plt.show()
```

Building 3d scatter plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
x = 20 * np.random.random(100)
y = np.sin(x)
z = np.cos(x)
ax.scatter(x,y,z, c=x, cmap='Blues')
plt.show()
```

# Plotting a Histogram

```python
plt.style.use('_mpl-gallery')

# Make data
np.random.seed(1)
x = 4 + np.random.normal(0,1.5,200)

# Plot
fig, ax = plt.subplots()

ax.hist(x, bins=8, linewidth=0.5, edgecolor='white')

ax.set(xlim=(0,8), xticks=np.arange(1,8), ylim=(0,56), yticks=np.linspace(0,56,9))

plt.show()
```
