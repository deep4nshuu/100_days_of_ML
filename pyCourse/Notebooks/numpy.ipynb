print(list1 + list2)

- But In Numpy arr, if we use +,-,* sign, it will do elementwise addn, subs, multiply of both arrays val in single arr

- And using .add, .sub, .multiply(matrix multi) will do opt done in maths

```python
# list
list1 = [1,2,3,4]
list2 = [5,4,7,8]
print(list1 + list2)
```

```python
# Np arr PROCESS 1
np1 = np.random.randint(0,10,(3,3))
print(np1)

np2 = np.random.randint(10,20,(3,3))
print(np2)

print(np1 + np2)
print(np1 - np2)
print(np1 * np2)
print(np1 / np2)
```

# Array Manipulatn

1. TRANSPOSE OF ARR

```python
arr = np.random.randint(0,10,(2,3))
print(arr)
print(arr.shape)
```

```python
# Process 1
trans = np.transpose(arr)
print(trans)
print(trans.shape)
```

```python
# Process 2
trans2 = arr.T
print(trans2)
print(trans2.shape)
```

2. RESHAPE : Changing dim of matrices

```python
a = np.random.randint(10,20,(3,2))
print(a)
print(a.shape)
```

```python
# Re-shaping a arr
b = a.reshape(2,3)
print(b)
print(b.shape)
```

> Difference b/w Transpose and Reshape

- In transpose, we replace first row to first col or vice-verse and so on for all other rows and cols.
- Also after two transpose we get same matrix.

- But in reshape, we take element sequentially or one by one from rows to fulfill ele ac to dim.
- After every operation, matrix get way different than original matrix

3. Flattened/ Raveled

```python
a = np.random.randint(10,20,(3,2))
print(a)

# Reshape
reshaped = a.reshape(2,3)
print(reshaped)

# Flatten
flattened = reshaped.flatten()
print("\n Flattened arr:",flattened)

# Ravel - Returns view instead of copy
raveled = reshaped.ravel()
print("\n Raveled arr", raveled)
```

4. INDEXING AND SLICING IN ARR

```python
# FOR 1D
arr = np.array([i for i in range(10)])
print(arr)
print("Basic slicing: ", arr[2:7])
print("With step: ", arr[2:7:2])
print("Negative Indexing : ", arr[-4])
```

```python
# For 2d
arr = np.array([(1,2,3),(9,8,7),(6,5,0)])
print("Specific idx: ", arr[2][2])
print("Entire row: ", arr[2])
print("Specific col: ", arr[:,2])
```

5. SORTING

AXIS :  0(For colwise orientation) & 1(For rowise orientation)

```python
# for 1d
unsort_arr = np.array([3,6,3,9,0,2,4,5])
print("Sorted arr: ", np.sort(unsort_arr))

# for 2d
unsort2d  = np.array([[3,2,1],[7,5,3],[9,6,0]])
print("2d sorted arr with col: ", np.sort(unsort2d, axis = 0))
print("2d sorted arr with row: ", np.sort(unsort2d, axis = 1))
```

6. FILTER

```python
# Numpy allows to write and execute expression in []
arr = np.array([3,6,3,9,0,2,4,5])
even_no = arr[arr % 2 == 0]   
print(even_no)
```

FILTER WITH MASK

```python
# Mask is a container of expression or store expression for evaluation which we can pass inside arr
mask = arr > 5
print("No greater than 5: ", arr[mask])
```

Fancy Indexing vs np.where()

```python
index = [2,3,6]
print(arr[index])  # it'll print giv idx val

# where stores particular val idx of arr
where_res = np.where(arr > 4)
print(where_res)
print(arr[where_res])
```

```python
# Special conditioning arr with multi condn or allow to evaluate multi-expression
condn_arr = np.where(arr > 5, arr * 5, arr) # it means no's who are greater than 5 should be multiplied by 5 and also print all no's
print(condn_arr)
```

```python
condn_arr = np.where(arr > 5, "True", "False")
print(condn_arr)
```

To add a new row to existing arr

```python
original = np.array([(1,2),(3,4)])

new_row = np.array([(5,6)])
with_new_row = np.vstack([original, new_row])

new_col = np.array([[7],[9]])
with_new_col = np.hstack([original, new_col])

print(original)
print(with_new_row)
print("With new col:",with_new_col)
```

```python
# to delete a ele 
arr = np.array([1,2,3,4,5,6])
deleted = np.delete(arr, 2)

print(deleted)
```

# Important tricK for idexing in arr using colon

[:, 1:]  --> this is divided into r&c, so for ist part having : means select all rows and for 2nd part, it defines range as colon(:) occur after 1 which means select from 1st col to last as end range is not written

```python
# DS: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 15000, 30000, 80000, 60000],  # Paradis Biryani
    [2, 20000, 35000, 20000, 30000],  # Beijing Bites
    [3, 20000, 60000, 50000, 45000],  # Pizza Hut
    [4, 18000, 10000, 15000, 40000],  # Burger Point
    [5, 17000, 90000, 20000, 60000],  # Chai points
])

print("\n Sample data for 1st 3 rest: ", sales_data[:3])
```

```python
# total sales per year
year_total = np.sum(sales_data[:,1:], axis=0)
print(year_total)
```

```python
# cumulative sum
cumsum = np.cumsum(sales_data[:,1:], axis=1)
print(cumsum)
```

```python
plt.figure(figsize=(10,10))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Avg cumsum sale per year")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
