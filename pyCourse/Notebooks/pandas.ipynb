# Manipulating a DataFrame

```python
# Adding a column to a DataFrame
diab_df['Price'] = db_dataset.target
```

```python
diab_df.head()
```

```python
# Removing a row
diab_df.drop(index=0, axis=0)
```

```python
# removing a col
diab_df.drop(columns='age', axis=1)
```

```python
# locating a row using idx val
diab_df.iloc[2]
```

```python
# locating particular col using idx
print(diab_df.iloc[:,0])  # first col
print(diab_df.iloc[:,1])
print(diab_df.iloc[:,2])
print(diab_df.iloc[:,-1])  # last col
```
