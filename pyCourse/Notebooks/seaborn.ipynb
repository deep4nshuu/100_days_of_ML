# Heat Map

It can be used to plot correlation vaues

Attributes:
- corr -> tell correlatn in value
- cbar -> color bar to show effect(high/low)
- square -> telling fig should be in square
- fmt -> floating pt no.(1f,2f,..)
- annot -> means x,y axis field name or annotations
- cmap -> color of map

```python
# constructing a Heat map
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Reds')
```

<h1 style='color:red'># USE CASE OF ALL TYPES OF PLOTS</h1>

| Plot            | Data Type   | Best For       |
| --------------- | ----------- | -------------- |
| plot / lineplot | Numerical   | Trend analysis |
| histogram       | Numerical   | Distribution   |
| kdeplot         | Numerical   | Density        |
| boxplot         | Numerical   | Outliers       |
| countplot       | Categorical | Frequency      |
| barplot         | Categorical | Comparison     |
| scatterplot     | Numerical   | Relationship   |
| pairplot        | Numerical   | EDA            |
| heatmap         | Matrix      | Correlation    |
| clustermap      | Matrix      | Clustering     |
| piechart        | Categorical | Proportion     |
| relplot         | Mixed       | Advanced plots |


<h1 style='color:green'> ALL PLOTS AND THIER USE CASE</h1>

<h3 style='color:brown'>1️⃣ plt.plot() (Line Plot – Matplotlib)</h3>

<p>Definition:</p>

- Used to show trends or changes in data over continuous intervals.

Syntax:

<ol>
<li>plt.plot(x, y)</li>
<li>plt.xlabel("X")</li>
<li>plt.ylabel("Y")</li>
<li>plt.show()</li>
</ol>


Use case:

<ul>
<li>✔ Stock prices</li>
<li>✔ Time-series data</li>
<li>✔ Trend analysis</li>
</ul>


Difference:

- Basic and low-level
- No automatic statistical handling (unlike seaborn lineplot)




<h4 style='color:brown'>2️⃣ Histogram</h4>

<p>Definition:</p>

- Shows frequency distribution of a numerical variable.

Syntax:

<ol>
<li>plt.hist(data, bins=10)</li>
<li>plt.show()</li>
</ol>


Use case:

<ul>
<li>✔ Data distribution</li>
<li>✔ Detect skewness</li>
<li>✔ Understand spread</li>
</ul>


Difference:

- Shows frequency
- No probability density (unlike KDE)




<h4 style='color:brown'>3️⃣ Distplot (Deprecated but conceptually important)</h4>

<p>Definition:</p>

- Combines histogram + KDE curve to show distribution.


<p>Syntax (older Seaborn):</p>

- sns.distplot(data)


Use case:

- ✔ Distribution + smooth density curve

Difference:

- Replaced by histplot() and kdeplot()

- Not recommended in new projects




<h4 style='color:brown'>4️⃣ Boxplot</h4>

<p>Definition:</p>

- Displays data distribution using quartiles and outliers.


<p>Syntax:</p>

- sns.boxplot(x=data)


<p>Use case:<p>

<ul>
<li>✔ Outlier detection</li>
<li>✔ Compare distributions</li>
<li>✔ Summary statistics</li>
</ul>


<p>Difference:</p>

- Shows median, Q1, Q3, outliers
- More statistical than histogram




<h4 style='color:brown'>5️⃣ Countplot</h4>

<p>Definition:</p>

- Shows count of categorical values.

Syntax:

- sns.countplot(x=category_column)


Use case:

<ul>
<li>✔ Categorical data frequency</li>
<li>✔ Survey analysis</li>
</ul>

Difference:

- For categorical data only
- Histogram is for numerical data




<h4 style='color:brown'>6️⃣ Pie Chart</h4>

<p>Definition:</p>

- Shows proportion or percentage of parts of a whole.

<p>Syntax:</p>

- plt.pie(values, labels=labels, autopct='%1.1f%%')
- plt.show()


<p>Use case:</p>

<ul>
<li>✔ Percentage contribution</li>
<li>✔ Market share</li>
</ul>


Difference:

- Not good for many categories
- Barplot is usually better




<h3 style='color:brown'>7️⃣ Scatter Plot</h3>

<p>Definition:</p>

- Shows relationship between two numerical variables.

<p>Syntax:<p>

- plt.scatter(x, y)
- plt.show()


<p>Use case:</p>

<ul>
<li>✔ Correlation analysis</li>
<li>✔ Pattern detection</li>
</ul>


Difference:

- No aggregation
- Each point represents one observation




<h3 style='color:brown'>8️⃣ Pairplot</h3>

<p>Definition:</p>

- Plots pairwise relationships for all numerical columns.

Syntax:

- sns.pairplot(df)


Use case:

<ul>
<li>✔ Exploratory Data Analysis (EDA)</li>
<li>✔ Correlation discovery</li>
</ul>


Difference:

- Automatic multi-plot
- Expensive for large datasets




<h3 style='color:brown'>9️⃣ Lineplot (Seaborn)</h3>

<p>Definition:</p>

- High-level line plot with confidence intervals.

<p>Syntax:</p>

- sns.lineplot(x=x, y=y)


<p>Use case:</p>

- ✔ Trend analysis with mean & CI

Difference:

- Automatically calculates mean
- Better than plt.plot() for datasets




<h3 style='color:brown'>🔟 Barplot</h3>

<p>Definition:</p>

- Shows comparison between categories using bars.

<p>Syntax:</p>

- sns.barplot(x=category, y=value)


<p>Use case:</p>

- ✔ Category comparison
- ✔ Average values

<p>Difference:</p>

- Uses mean by default
- Countplot shows frequency




<h3 style='color:brown'>1️⃣1️⃣ Heatmap</h3>

Definition:

- Represents values using color intensity.

Syntax:

- sns.heatmap(data, annot=True)


<p>Use case:</p>

- ✔ Correlation matrix
- ✔ Pattern visualization

Difference:

- Matrix-based
- Color-encoded values




<h3 style='color:brown'>1️⃣2️⃣ Clustermap</h3>

Definition:

- Heatmap with hierarchical clustering.

Syntax:

- sns.clustermap(data)


Use case:

- ✔ Cluster detection
- ✔ Advanced data analysis

Difference:

- Adds dendrograms
- Computationally heavier than heatmap



<h3 style='color:brown'>1️⃣3️⃣ KDE Plot (kdeplot)</h3>

Definition:

- Shows smooth probability density of data.

Syntax:

- sns.kdeplot(data)

Use case:

- ✔ Understand data distribution
- ✔ Density estimation

Difference:

- Smooth curve
- No frequency bars like histogram



<h3 style='color:brown'>1️⃣4️⃣ Relplot</h3>

Definition:

- Figure-level interface for scatter & line plots.

Syntax:

- sns.relplot(x=x, y=y, kind="scatter", data=df)


Use case:

- ✔ Multi-facet relational plots
- ✔ Complex comparisons

Difference:

- High-level wrapper
- Internally uses scatterplot or lineplot
