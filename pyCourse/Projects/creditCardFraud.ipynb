```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data = pd.read_csv(r'C:\Users\HP\OneDrive\Desktop\python\pyCourse\csvFiles\credit_data.csv')
```

```python
credit_card_data.head()
```

- Info about dataset

```python
credit_card_data['Class'].value_counts()
```

- 0  --> Legit
- 1  --> Fraud

```python
credit_card_data.describe()
```

- distributing transctn

```python
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape, fraud.shape)
```

```python
credit_card_data.groupby('Class').mean()
```

- Under sampling

```python
legit_sample = legit.sample(n=500)
```

- Concatenate sample and fraud trans

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.groupby('Class').mean()
```

Here, mean data of both(creditcard & newData) is similar, So it's good sample dataset

- Separating feature and target

```python
x = new_dataset.drop(columns='Class', axis=1)
y = new_dataset['Class']
```

```python
print(y)
```

- Splitting data into train & test

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

- Model Training

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

- Evaluation

```python
x_train_pred = model.predict(x_train)
training_pred = accuracy_score(x_train_pred, y_train)
```

```python
print(training_pred)
```

```python
x_test_pred = model.predict(x_test)
testing_pred = accuracy_score(x_test_pred, y_test)
```

```python
print(testing_pred)
```
